```python
X_train, y_train=df_train['story'],df_train['gender']
```

```python
def CrossVal(X_train,y_train,grid_search,model):
    print("Model:", model)
    scores =cross_val_score(grid_search.best_estimator_, X_train, y_train, scoring=make_scorer(f1_score,average='macro'), cv=10)
    average_f1=scores.mean()
    print("Average_f1 score: ", average_f1)
    return average_f1,grid_search.best_params_,grid_search.best_estimator_
```

```python
def GridSearchModel(model,X_train,y_train ):
    # Define the parameter grid to search over
    # Create the diffrent models
    if model=='LinearSVC':        # Create the LinearSVC model
        param_grid = {
        'C': [0.01, 0.1, 1.0, 10.0, 100.0],
        'class_weight': [None, 'balanced'] }
        clf=LinearSVC()
    
    if model=='Perceptron':         # Create the Perceptron model
        param_grid = {
        'penalty': ['l1', 'l2', 'elasticnet', 'none'],
        'alpha': [0.0001, 0.001, 0.01, 0.1, 1],
        'max_iter': [100, 1000, 10000],
        'eta0': [0.0001, 0.001, 0.01, 0.1, 1] }
        clf = Perceptron()
    
    if model=='SGDClassifier':        # Create the SGDClassifier model
        param_grid = {
        'alpha': [0.0001, 0.001, 0.01, 0.1],
        'loss': ['hinge', 'log', 'modified_huber'],
        'penalty': ['l2', 'l1', 'elasticnet'] }
        clf = SGDClassifier()

    if model=='MultinomialNB':          # Create a MultinomialNB classifier
        param_grid = {
        'alpha': [ 0.1, 1, 10, 100],
        'fit_prior': [False,True]}
        clf = MultinomialNB()
    
    if model=='KNeighborsClassifier':        #Create the KNeighborsClassifier model
        param_grid = {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan','Chebyshev'] }
        clf = KNeighborsClassifier()
        
    grid_search = GridSearchCV(clf, param_grid=param_grid, cv=10, scoring='f1_macro')
    average_f1,params ,best_estimator = CrossVal(X_train,y_train,grid_search.fit(X_train, y_train),model)
    return average_f1 , params , clf, best_estimator
```

```python
# Choose whether to use the CountVectorizer or TfidfVectorizer
def ChoseVector(X_train,vec,ngram,mindf,max_features):
    if vec=='CountVectorizer':
        vector=CountVectorizer(ngram_range=ngram, max_features= maxfeatures ,min_df=mindf)
        
    if vec=='TfidfVectorizer':
        vector=TfidfVectorizer(ngram_range=ngram, max_features= maxfeatures ,min_df=mindf)
        
    Vec_X_train  = vector.fit_transform(X_train) 
    print("Using min_df ",mindf,"\nUsing max_features ",max_features,"\nUsing ngram",ngram)
    return vector,Vec_X_train
```

```python
# Mix the different models using different ngrams and different tokens
#Show all the different results and scores
MaxScore=0
for vec in ['CountVectorizer','TfidfVectorizer']:
    print('--------------' + vec + '--------------')
    for mindf in[2,3,4]:
        for maxfeatures in [14500,15000,15500]:
            for ngram in [ (1, 2), (1, 3), (1, 4),  (2, 3), (2,4)]:
                vectorizer ,Vec_X_train = ChoseVector(X_train,vec,ngram,mindf,maxfeatures)
                for model in ['KNeighborsClassifier','MultinomialNB','SGDClassifier','Perceptron','LinearSVC']:
                    average_f1,Param,clf,estimator= GridSearchModel(model,Vec_X_train,y_train)
                    print("----------------------------")
                    if(average_f1>MaxScore):
                        print("Best model is changed!")
                        print("----------------------------")
                        BestClf=clf
                        MaxScore= average_f1
                        BestParam=Param
                        CorrectVector=vectorizer
                        best_estimator=estimator
    
```
